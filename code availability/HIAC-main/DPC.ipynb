```python
file_path = r"D:\HRG\Clustering\DP\code availability\DensityPeakCluster-master lanbing510\data\data_others\aggregation.txt"
dataset = np.loadtxt(file_path)
label = dataset[:, -1].astype('i')
data = dataset[:, :2]
```
